```python
import pandas as pd
import numpy
from collections import Counter

from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import Ridge
%pylab inline
```

```python
def situation(row):
    if row['position'] == 0:
        return "first"
    elif row['position'] == 1:
        return "second"
    elif row['position'] == 2:
        return "third"
    elif row['position'] + 1 == row['length']:
        return "last"
    elif row['position'] + 1 == row['length'] - 1:
        return "penult"
    elif row['position'] + 1 == row['length'] - 2:
        return "antepenult"
    else:
        return "middle"
        
def read_data(path):
    data = pd.read_csv(path, header=False, 
                       names=["id","id2","word","pos","deprel","score"])[["id","word","pos","deprel","score"]]
    position = []
    length = []
    i = -1
    for word in data["word"]:
        if word=="<END>":
            length.append(-1)            
            for j in range(1,i+3):
                length[-j] = i+1
            i = -1
            position.append(i)
        else:
            i += 1
            position.append(i)
            length.append(-1)
    data["position"] = position
    data["length"] = length
    data["situation"] = data.apply(situation, axis=1)
    data = data[data["word"]!="<END>"].dropna()
    data["wordxdeprel"] = data["word"]+":"+data["deprel"]
    data["wordxsituation"] = data["word"]+":"+data["situation"]
    return data


def filter_by_count(df, col, count):
    freqs = Counter(df[col])
    counts = df[col].apply(lambda x: freqs[x])
    return df[counts > count]

def prepare_data_old(data):
    encoder = {}
    encoder['deprel_map'] = LabelEncoder()
    encoder['pos_map']    = LabelEncoder()
    encoder['word_map']   = LabelEncoder()
    encoder['situation_map'] = LabelEncoder()
    encoder['wordxdeprel_map'] = LabelEncoder()
    encoder['wordxsituation_map'] = LabelEncoder()
    
    data.loc[:, "deprel_id"] = encoder['deprel_map'].fit_transform(numpy.array(data["deprel"]))
    data.loc[:, "pos_id"]    = encoder['pos_map'].fit_transform(numpy.array(data["pos"]))
    data.loc[:, "word_id"]   = encoder['word_map'].fit_transform(numpy.array(data["word"]))
    data.loc[:, "situation_id"] = encoder['situation_map'].fit_transform(numpy.array(data["situation"]))
    data.loc[:, "wordxdeprel_id"] = encoder['wordxdeprel_map'].fit_transform(numpy.array(data["wordxdeprel"]))
    data.loc[:, "wordxsituation_id"] = encoder['wordxsituation_map'].fit_transform(numpy.array(data["wordxsituation"]))
    
    X_deprel = OneHotEncoder().fit_transform(numpy.array(data[["word_id","deprel_id","wordxdeprel_id"]]))
    X_situation = OneHotEncoder().fit_transform(numpy.array(data[["word_id","situation_id", "wordxsituation_id"]]))
    X_full = OneHotEncoder().fit_transform(numpy.array(data[["word_id","deprel_id","wordxdeprel_id",
                                                             "situation_id", "wordxsituation_id"]]))
    X_word = OneHotEncoder().fit_transform(numpy.array(data[["word_id"]]))
    Y = numpy.array(data["score"])
    return dict(full=X_full, word=X_word, deprel=X_deprel,
                situation=X_situation, Y=Y, encoder=encoder)

def ridge_alphas(D, ix=100000):
    Y = D['Y']
    scores = dict(word=[], deprel=[], situation=[], full=[])
    alphas = [ 2**n for n in range(-2,6) ]
    
    def score(alpha, modeltyp):
        model = Ridge(alpha=alpha)
        model.fit(D[modeltyp][:ix,:], Y[:ix])
        scores[modeltyp].append(model.score(D[modeltyp][ix:,:], Y[ix:]))
        
    for alpha in alphas:
        for modeltyp in ['word','deprel','situation','full']:
            score(alpha, modeltyp)
    return (alphas, scores)
```

```python
data_viz = read_data("../data/visual_full_omission.csv")
data_viz.to_csv("visual_omission_with_position.csv", index=False)

data_txt = read_data("../data/textual_full_omission.csv")
data_txt.to_csv("textual_omission_with_position.csv", index=False)

data_sum = read_data("../data/sum_visual_omission.csv")
data_sum.to_csv("sum_visual_omission_with_position.csv", index=False)
```

```python
def choose(ds, keys):
        return  ( dict([(key,d[key]) for key in keys ]) for d in ds )

def prepare_data(data):   
    dicts = data.T.to_dict().values()
    encoder = dict(word=DictVectorizer(), deprel=DictVectorizer(), situation=DictVectorizer(), full=DictVectorizer())
    X_word   = encoder['word'].fit_transform(choose(dicts, ['word']))
    X_deprel = encoder['deprel'].fit_transform(choose(dicts, ['word', 'deprel', 'wordxdeprel']))
    X_situation = encoder['situation'].fit_transform(choose(dicts, ['word','situation','wordxsituation']))
    X_full = encoder['full'].fit_transform(choose(dicts, ['word','deprel','wordxdeprel','situation','wordxsituation']))
    return dict(Y=numpy.array(data['score']), encoder=encoder, word=X_word, deprel=X_deprel, situation=X_situation, full=X_full)
    
```

```python
D_viz = prepare_data(data_viz)
D_sum = prepare_data(data_sum)
D_txt = prepare_data(data_txt)
```

```python
alphas, scores_viz = ridge_alphas(D_viz)
alphas, scores_sum = ridge_alphas(D_sum)
alphas, scores_txt = ridge_alphas(D_txt)
```

```python
def plot_alphas(alphas, scores):
    figure(figsize=(7.0, 5.0))
    matplotlib.rcParams.update({'font.size': 15})
    subplot()
    semilogx(alphas, scores['word'], label="word")
    semilogx(alphas, scores['deprel'], label="deprel")
    semilogx(alphas, scores['situation'], label="position")
    semilogx(alphas, scores['full'], label="full")
    xlabel("alpha")
    ylabel(r"Heldout $R^2$")
    legend(loc="best")
    savefig("scores-vs-alpha.pdf")
```

```python
def score_diffs(alpha_word, alpha_full, D, data, ix=100000):
    X_word = D['X_word']
    X_full = D['X_full']
    Y = D['Y']
    
    reg_word = Ridge(alpha=alpha_word)
    reg_full = Ridge(alpha=alpha_full)

    reg_word.fit(X_word[:ix,:], Y[:ix])
    reg_full.fit(X_full[:ix,:], Y[:ix])
    pred_word = reg_word.predict(X_word)
    pred_full  = reg_full.predict(X_full)
    err_word = numpy.abs(pred_word - Y)
    err_full = numpy.abs(pred_full - Y)
    err_diff = err_full - err_word
    data.loc[:,"err_diff"] = err_diff
    data.loc[:,"pred_word"] = pred_word
    data.loc[:,"pred_full"] = pred_full
    ordered = data.iloc[ix:,:].sort(columns=["err_diff"])\
        [["word","deprel","pos","score","pred_word","pred_full","err_diff"]]
    return ordered

def rank_words(ordered):
    return filter_by_count(ordered,"word",100).groupby(["word"])[["word","err_diff"]].mean().sort(columns=["err_diff"])
    
```
